```python
# Histogram of Age
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
### Observation

- Most passengers were between **20 and 40 years** of age.
- There were relatively few elderly passengers.
- The age distribution is slightly right-skewed.
```

```python
# Boxplot of Fare
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Fare"])
plt.title("Boxplot of Fare")
plt.show()
```

```python
### Observation

- The fare distribution contains several **outliers**.
- Most passengers paid relatively low fares.
- A few passengers paid exceptionally high ticket prices.
```

```python
# Scatter Plot: Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
### Observation

- There is no strong relationship between passenger age and ticket fare.
- Most passengers paid lower fares regardless of age.
- A few passengers paid exceptionally high fares (outliers).
```

```python
# Pair Plot
sns.pairplot(df[["Age", "Fare", "Pclass", "Survived"]])
plt.show()
```

```python
### Observation

- Fare varies significantly across passengers.
- Most passengers belong to Passenger Class 3.
- No strong linear relationship exists among the numerical variables.
- Pairplot helps visualize distributions and relationships simultaneously.
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
### Observation

- Fare shows a moderate negative correlation with Passenger Class (Pclass).
- Passenger Class has a negative correlation with Survived, indicating that first-class passengers had better survival rates.
- Most other numerical variables have weak correlations.
- The heatmap helps identify relationships between numerical features.
```

```python
# Summary of Findings

- The dataset contains 891 passenger records with both numerical and categorical features.
- Missing values are present in the Age, Cabin, and Embarked columns.
- Most passengers were between 20 and 40 years of age.
- The majority of passengers traveled in Third Class.
- Male passengers outnumbered female passengers.
- More passengers did not survive than survived.
- Ticket fares contained several outliers, with a few passengers paying significantly higher fares.
- No strong relationship was observed between passenger age and ticket fare.
- Correlation analysis showed only a few moderate relationships among numerical variables.

## Conclusion

The exploratory data analysis provided useful insights into passenger demographics, ticket pricing, 
survival distribution, and feature relationships. Visualizations helped identify trends,
missing values, and outliers, making the dataset easier to understand before any further 
analysis or predictive modeling.
```
